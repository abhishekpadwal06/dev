Plotting with Categorical Data

catplot [Figure Level Function]
Just like relplot(), it can utilize other access level functions to draw different kinds of graphs


1. Categorical Scatterplots
    - Stripplot
    - Swarmplot

2. Categorical distribution plots
    - Boxplot
    - Violinplot

3. Categorical estimate plots
    - Pointplot
    - Barplot
    - Countplot

4. Categorical categorical plots
    - Heatmaps
    - Clustermaps

5. Plotting side-by-side graphs using FacetGrids

```python
# x - axis ==> Categorical Data
# y - axis ==> Numerical Data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.catplot(x='day', y='tip', kind='strip', data=data) # By default kind='strip' 
```

```python
sns.catplot(x='day', y='tip', kind='swarm', data=data)
```

```python
# Instead of catplot(), even if we do swarmplot(), we will get the same chart
sns.swarmplot(x='day', y='tip', data=data)
```

```python
# We can use the 'hue' parameter here as well!
sns.catplot(x='day', y='tip', kind='swarm', hue='sex', data=data)
```
