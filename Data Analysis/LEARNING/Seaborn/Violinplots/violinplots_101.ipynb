```python
# Violin Plots - Combination of boxplot with kernel density plot (basically PDF (probability density function))

# Advantage - Also shows the entire distribution of data
# Disadvantage - Slightly harder to read than boxplots
```

```python


import seaborn as sns

```

```python
data = sns.load_dataset('tips')
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=data)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', data=data)
```

```python
# Since this looks to cluttered, you can use a paramter called 'split'

sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', split=True, data=data)
```
