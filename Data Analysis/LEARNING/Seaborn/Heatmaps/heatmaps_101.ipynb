Heatmaps - 
 - A graphical representation of data where the individual values contained in a matrix are represented as colours

Generally drawn between 2 categorical datas or sometimes numerical data as well 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = sns.load_dataset('flights')
data.head()
```

```python
x = data.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
# sns.heatmap(x, annot=True, fmt='d', cmap='summer')
# sns.heatmap(x, annot=True, fmt='d', cmap='winter')
sns.heatmap(x, annot=True, fmt='d', cmap='viridis')
```
