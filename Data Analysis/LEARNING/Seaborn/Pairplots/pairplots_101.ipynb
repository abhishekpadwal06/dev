Pairplots - 

These are used to plot multiple bivariate distributions in a dataset. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
data = sns.load_dataset("iris")
data.head()
```

```python
sns.pairplot(data)

# Basically what it has done is, it has figured out pair-wise relationship with all the columns of the entire DataFrame.
# Now, if 2 columns are same it is univariate (Hence, a histogram is plotted)
# Else it is bivariate (Hence, a scatterplot is plotted)
```

```python
# Using the 'hue' parameter
sns.pairplot(data, hue='species')

# Now here, all the univariate analysis (histograms) are replaced by PDFs, and everything is colour-coded!
```
