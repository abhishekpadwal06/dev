Boxplot - (Very much similar to a PDF (probability distribution function))

Standardized way of displaying the distribution of data based on a five number summary

(minimum, first quartile (Q1), median, third quartile (Q3) and maximum). It also tells you about the outliers and what their values are! 

Also tells you if data is symmetrical, how tightly your data is grouped, and if and how your data is skewed



Terms in a Box Plot

1. Minimum
    The smallest value in the dataset excluding outliers.



2. First Quartile (Q1) - 25th Percentile
    25% of the data lies below Q1.



3. Median (Q2) - 50th Percentile
    The middle value of the dataset.
        50% of values are below it.
        50% are above it.



4. Third Quartile (Q3) - 75th Percentile
    75% of the data lies below Q3.



5. Interquartile Range (IQR)
    Measures the spread of the middle 50% of data.
        IQR=Q3−Q1



6. Whiskers
    Lines extending from the box to the smallest and largest non-outlier values.



7. Outliers
    Values that are unusually far from the rest of the data.

    A common rule:
        Lower Bound=Q1−1.5×IQR
        Upper Bound=Q3+1.5×IQR

    Any value outside these bounds is plotted as an outlier.
