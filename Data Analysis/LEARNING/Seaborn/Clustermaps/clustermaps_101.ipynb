Clustermaps - 

A Cluster Map (Clustermap) is a heatmap that automatically groups similar rows and columns together using clustering algorithms.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = sns.load_dataset('flights')
data.head()
```

```python
x = data.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.clustermap(x)
```
