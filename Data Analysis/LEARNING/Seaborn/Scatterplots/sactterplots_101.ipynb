Seaborn - 
Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics with less code.

Why use Seaborn?
1. Easier to create complex visualizations than Matplotlib.
2. Better default styles and color themes.
3. Works seamlessly with Pandas DataFrames.
4. Includes functions for statistical plotting and data exploration.

Relational / Statistical Plots
 - Used to perform multivariate analysis

Types - 
1. Scatter Plots
2. Line Plots (Specifically used for time series data)
3. FacetGrids (Plotting multiple graphs side-by-side)

Two ways to plot a scatterplot - 
1. relplot() - [Figure Level Function]
2. scatterplot() - [Axes Level Function]

Preferred method (relplot) Because you can plot both Axes level and Figure level using relplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Loading the dataset
data = sns.load_dataset('tips')         # This is a built-in dataset in seaborn!
data.head()
```

```python
# Plot a scatter plot between 'total_bill' and 'tip' 
sns.relplot(x='total_bill', y='tip', kind='scatter', data=data)
```

```python
# Now let's plot the same scatterplot again but this time we will see one more analysis
# i.e., whether the tip given is given by a smoker or not!

# For that, we will use the 'hue' parameter (Enter a categorical data)

sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
```

```python
# Now let's see whether the person who tipped 2$ for a 30$ bill was a guy / girl

# This can be done using the 'style' parameter (Enter a categorical data)

sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', data=data)
```

```python
# 'size' parameter (Enter a numerical / discrete data)

sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', size='size', data=data)
```

```python
# Using scatterplot()

sns.scatterplot(x='total_bill', y='tip', hue='smoker', style='sex', size='size', data=data)
```
