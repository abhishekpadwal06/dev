Histogram Plot (histplot) - 

It is used to visualize the distribution of a numerical variable. It shows how frequently different ranges of values occur in your data.


--- Important Terms ---

1. Bins - Intervals used to group data.
2. Frequency - The height of each bar represents how many observations fall in that bin.
3. KDE (Kernel Density Estimate) - A smooth curve showing the distribution shape.
4. Density - Instead of counts, you can show probabilities.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
# Handling null values

data['age'] = data['age'].fillna(data['age'].mean())
```

```python
# Plotting a histplot

sns.histplot(data['age'])
```

```python
sns.histplot(data['age'], kde=True)
```

```python
sns.histplot(data['age'], kde=True, stat='density')

# Instead of counts, it shows probabilities.
```
