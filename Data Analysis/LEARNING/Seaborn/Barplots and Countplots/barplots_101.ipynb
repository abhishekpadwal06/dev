Barplots - 
These give information about central tendency rather than distribution of data

Let's find minimum tip for smokers and non-smokers

```python
import seaborn as sns
import numpy as np
```

```python
data = sns.load_dataset("tips")
data.head()
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', data=data)

# Black bars in the image are 'error bars' and they are actually showing confidence interval in calculating mean!
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', data=data)
```

```python
# Optional parameter - estimator

# Now basically barplots are used to calculate mean of any numerical data, but let's say we want to calculate median, mode, etc., 
# then in that case, we will use the estimator parameter

sns.catplot(x='smoker', y='total_bill', estimator=np.median, kind='bar', hue='sex', data=data)
```

Countplots - 
Basically a special case of barplots. 
Similar to histograms.
Used for finding count of categorical value.

```python
sns.catplot(x='sex', hue='smoker', kind='count', data=data)
```
