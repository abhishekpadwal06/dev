```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data)
```

```python
# Hexplot - Basically a 2D histogram


sns.jointplot(x='total_bill', y='tip', kind='hex', data=data)
```

```python
# 2D - KDE plots

sns.jointplot(x='total_bill', y='tip', kind='kde', data=data)
```

```python
# Regression Plot

sns.jointplot(x='total_bill', y='tip', kind='reg', data=data)


# It has created a line called the 'Best fit Line'. Basically if we get a new total_bill, we can plot it's corresponding tip! 
# or vice-versa
```
