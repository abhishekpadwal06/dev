Subplot

Figure  - container for holding our plots
Axes    - actual plots

Therefore a figure can have multiple plots!

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, py_salaries, label='Python', linewidth=2)
plt.plot(ages, js_salaries, label='JavaScript', linewidth=2)
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs', linewidth=2)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
plt.style.use("bmh")
data = pd.read_csv('data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, ax = plt.subplots()
ax.plot(ages, py_salaries, label='Python', linewidth=2)
ax.plot(ages, js_salaries, label='JavaScript', linewidth=2)
ax.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs', linewidth=2)

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
# After doing all that we are still getting the same chart but only change is that we have made a new axis  
# Now what's the use of that new axis is that let's say we want Python and JS plots together on one plot 
# But we want the 'All devs' plot on a seperate plot, that's when this new axis becomes helpful
```

```python
plt.style.use("bmh")
data = pd.read_csv('data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=True)

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs', linewidth=2)
ax2.plot(ages, py_salaries, label='Python', linewidth=2)
ax2.plot(ages, js_salaries, label='JavaScript', linewidth=2)

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

fig.savefig('fig.png')
```

```python
data = pd.read_csv('data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs', linewidth=2)
ax2.plot(ages, py_salaries, label='Python', linewidth=2)
ax2.plot(ages, js_salaries, label='JavaScript', linewidth=2)

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```
