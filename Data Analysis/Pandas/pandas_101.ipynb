```python
# Now let's say we want to categorize athletes into short, avg and tall.
# So to do that, what we gotta do is - 
bios_new['Height_Category'] = bios_new['height_cm'].apply(lambda x: 'Short' if x < 165 else('Average' if  x < 185 else 'Tall'))
```

```python
# Now what we did above the same can be done by defining functions 
def categorize_athletes(row):
    if row['height_cm'] < 165 and row['weight_kg'] <= 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 85:
        return 'Middleweight'
    else:
        return 'Heavyweight'
    
bios_new['Category'] = bios.apply(categorize_athletes, axis=1)  # 1 -> rows, 0 -> columns
```

Handling Null Values

```python
coffee_new.head()
```

```python
# If you want to explicitly add NaN (Not-a-Number) in your df, then -->
coffee_new.loc[[2, 3], 'Units Sold'] = np.nan
```

```python
coffee_new.head()
```

```python
coffee_new.info()
```

```python
coffee_new.head()
```

```python
# We can handle null values using multiple things:

# 1. Using mean()
coffee_new.fillna(coffee_new['Units Sold'].mean())

# 2. Using interpolate(), meaning it sees if there's a pattern (checks neighbours) and then accordingly it continues that pattern in the column
coffee_new['Units Sold'] = coffee_new['Units Sold'].interpolate()

# 3. Or we can simply drop those rows!
# coffee_new.dropna()
```

```python
coffee_new.head()
```

```python
# If you want to view all the rows having an NaN value in a particular column - 
coffee_new[coffee_new['Units Sold'].isna()]

# If you want to view all the rows NOT having an NaN value in a particular column - 
coffee_new[coffee_new['Units Sold'].notna()]
```

Advanced Functionality

```python
coffee_new.head()
```

```python
# shift()   -   Moves data up(backwards) / down(forward) by specified number of rows
# If you want to move up (backwards) enter a negative number in the shift()

# Comparing yesterday's and today's revenue of Latte and Espresso!
coffee_new["Yesterday's revenue"] = coffee_new['revenue'].shift(2)
```

```python
coffee_new.tail()
```

```python
# rank()    -   Assigns a rank (position) to each value based on its order.

# Rank all the athletes based upon their height!
bios_new['height_rank'] = bios_new['height_cm'].rank(ascending=False)       # Tallest person gets first rank!
bios_new.sort_values(['height_rank'])[['first_name', 'height_cm', 'height_rank']].head()
```

```python
# rolling() -   Creates a moving window over data and performs calculations line mean, min, max, or sum
# cumsum()  -   Keeps on adding new values to the previous sum


# Find the no of latte's sold for the past 3 days only!
latte = coffee_new[coffee_new['Coffee Type'] == 'Latte'].copy()
latte['3Day_Coffees Sold'] = latte['Units Sold'].rolling(3).sum()
```
