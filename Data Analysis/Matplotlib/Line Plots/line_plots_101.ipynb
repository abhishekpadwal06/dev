Line Plots using Fills (fill betweens)

```python
# Simple Line graph - 

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')

ages = data['Age']

dev_sal = data['All_Devs']
plt.plot(ages, dev_sal, label='Developers')

py_sal = data['Python']
plt.plot(ages, py_sal, label='Python', linestyle='--')

plt.legend()
plt.title("Salary (USD) by Age for various developers")
plt.xlabel("Ages")
plt.ylabel("Yearly salaries")
plt.show()
```

```python
# Now using fill betweens - 

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')

ages = data['Age']

dev_sal = data['All_Devs']
plt.plot(ages, dev_sal, label='Developers')

py_sal = data['Python']
plt.plot(ages, py_sal, label='Python', linestyle='--')


# Now let's say we wanna find out when exactly the median salary was crossed? (At what age?) --> 
# For that use 'y2' value, by default it is set to zero, meaning it will fill everything till the end (bottom)
overall_median = 57287
plt.fill_between(ages, dev_sal, overall_median, alpha=0.25)

plt.legend()
plt.title("Salary (USD) by Age for various developers")
plt.xlabel("Ages")
plt.ylabel("Yearly salaries")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')

ages = data['Age']

dev_sal = data['All_Devs']
plt.plot(ages, dev_sal, label='Developers')

py_sal = data['Python']
plt.plot(ages, py_sal, label='Python', linestyle='--')

overall_median = 57287
plt.fill_between(ages, dev_sal, overall_median, where=(dev_sal > overall_median), interpolate=True, alpha=0.25)
plt.fill_between(ages, dev_sal, overall_median, where=(dev_sal <= overall_median), interpolate=True, alpha=0.25)

plt.legend()
plt.title("Salary (USD) by Age for various developers")
plt.xlabel("Ages")
plt.ylabel("Yearly salaries")
plt.show()
```

```python
# Filling the area between 2 salaries -
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')

ages = data['Age']

dev_sal = data['All_Devs']
plt.plot(ages, dev_sal, label='Developers')

py_sal = data['Python']
plt.plot(ages, py_sal, label='Python', linestyle='--')

overall_median = 57287
plt.fill_between(ages, dev_sal, py_sal, where=(py_sal > dev_sal), interpolate=True, alpha=0.25, label='Above avg')
plt.fill_between(ages, dev_sal, py_sal, where=(py_sal <= dev_sal), interpolate=True, alpha=0.25, label='Below avg')

plt.legend()
plt.title("Salary (USD) by Age for various developers")
plt.xlabel("Ages")
plt.ylabel("Yearly salaries")
plt.show()
```
