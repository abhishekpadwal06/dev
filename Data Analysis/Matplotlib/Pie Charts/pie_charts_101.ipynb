Pie Charts

```python
from matplotlib import pyplot as plt
```

```python
slices = [30, 50, 30, 96, 131]       # Values need not add upto 100%. They can be anything! They will still take up the same proportions of the pie chart
labels = ['Thirty', 'Fifty', 'Thirty', 'Extra1', 'Extra2']
colors = ['Black', 'Red', 'Blue', 'Yellow', 'Grey']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'white'})

plt.title("My first Pie Chart")
plt.show()
```

```python
plt.style.use('fivethirtyeight')
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})

plt.title("My first Pie Chart")
plt.show()

# Plotting these many items in a pie chart looks shabby, prefer bar graphs over these!
```

```python
plt.style.use('fivethirtyeight')
# Exploding a particular value out of the pie chart, for emhasis purpose

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.2, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={'edgecolor':'white'}, startangle=90, autopct='%1.1f%%')

plt.title("My first Pie Chart")
plt.tight_layout()
plt.show()
```
