```python
# Real World Example - 

data = pd.read_csv('data.csv')
price_date = data['Date']
price_close = data['Close']

plt.plot(price_date, price_close, marker='o')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```

```python
# The thing is that, now the dates that we are getting are in string format and not in datetime format
# To prove this what we will do is copy the top line from the csv file and paste it in the end and rename that newly added line to May 17
# Now if it were in datetime format it should've plotted the data in the very beginning 
# But instead it is still plotted in the end 
# This proves that it is not in the datetime format!


data = pd.read_csv('data.csv')
price_date = data['Date']
price_close = data['Close']

plt.plot(price_date, price_close, marker='o')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```

```python
# Now let's change into datetime format!

data = pd.read_csv('data.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot(price_date, price_close, marker='o')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```
